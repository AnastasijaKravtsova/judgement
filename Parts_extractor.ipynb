```python
def clean(text):
    
    pattern1 = re.compile(r'(п|ч|ст)(\.|\s|\d)')
    pattern2 = re.compile(r'((У|Г)П?К|КоАП|ПДД)\sРФ')
    pattern3 = re.compile('\d')
    pattern4 = re.compile('(ДД.ММ.ГГ|дд.мм.гг)')

    text = re.sub(pattern1, '', text)
    text = re.sub(pattern2, '', text)
    text = re.sub(pattern3, '', text)
    text = re.sub(pattern4, '', text)
    text = text.replace('\xad', '')
    text = text.replace('п р и г о в о р и л', '')
    text = text.replace('\ufeff', '')
    text = text.replace('\ufeff', '')
    text = text.replace('п о с т а н о в и л', '')
    text = text.replace('П Р И Г О В О Р И Л', '')
    text = text.replace('П О С Т А Н О В И Л', '')
    
    return text
```

```python
def get_part(text, start_point, end_point):
    beg = text.split(start_point)[1:]
    res = [elem.split(end_point)[0] for elem in beg]
    return res
```

```python
meditations = []
fabulas = []
proves = []
witnesses = []

for text in judices_texts:
    meditations.extend(get_part(text, 'meditationb', 'meditatione'))
    fabulas.extend(get_part(text, 'fabb', 'fabe'))
    proves.extend(get_part(text, 'proofb', 'proofe'))
    witnesses.extend(get_part(text, 'witnessb', 'witnesse'))
    
    
print(len(meditations), len(fabulas), len(proves), len(witnesses))
```

```python
def remove_trash(text):
    trash = r'\n|ACCUSED|[.,;?!\'№:():-]|\\xa0|ADVOCATE|JUDGE|SECRETARY|PROSECUTOR|pleaguilty|condb|conde|\xa0|\\\\'
    no_trash_text = re.sub(trash, ' ', text)
    no_space_sequences_text = re.sub(' +', ' ', no_trash_text)
    return no_space_sequences_text
```
